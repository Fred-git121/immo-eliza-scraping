```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import csv

# URL listesi
url_list = [
    "https://immovlan.be/en/detail/apartment/for-sale/1060/sint-gillis/vwd15254",
    "https://immovlan.be/en/detail/apartment/for-sale/1000/brussels/rwc41104","https://immovlan.be/en/detail/villa/for-sale/1950/kraainem/vbd48547"
]

# Chrome ba≈ülat
driver = webdriver.Chrome()

# CSV dosyasƒ±nƒ± a√ß
with open("houses.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow([
        "Property ID","Locality name","Postal code","Price", 
        "Type of property", "Subtype of property",
        "Number_of_rooms", "Living area","Equipped kitchen", "Furnished", 
        "Terrace", "Garden", "Number of facades","Swimming pool","State of building"
    ])

    for url in url_list:
        driver.get(url)
        time.sleep(3)

        # Cookie izni
        try:
            driver.find_element(By.ID, "didomi-notice-agree-button").click()
            time.sleep(1)
        except:
            pass

        # Property ID
        Property_ID = driver.find_element(By.XPATH, '//*[@id="main_content"]/div[1]/div[1]/div/h1/span/span[2]').text.strip()
        Locality_name = driver.find_element(By.XPATH, '//*[@id="main_content"]/div[1]/div[1]/div/h1/span').text.strip()
        Postal_code = driver.find_element(By.XPATH, '//*[@id="main_content"]/div[1]/div[1]/div/div[2]/div/span[2]').text.strip()
        price = driver.find_element(By.XPATH, '//*[@id="main_content"]/div[1]/div[1]/div/p/span[1]').text.strip()

        # Type & Subtype
        title_text = driver.find_element(By.XPATH, '//*[@id="main_content"]/div[1]/div[1]/div/h1/span').text.lower()
        property_type = "Other"
        subtype = "Other"
        if any(word in title_text for word in ["apartment", "flat", "studio", "duplex"]):
            property_type = "Apartment"
        elif any(word in title_text for word in ["house", "villa", "bungalow", "residence", "farm", "manor", "mansion"]):
            property_type = "House"
        for word in ["apartment", "flat", "studio", "duplex", "house", "villa", "bungalow", "residence", "farm", "manor", "mansion"]:
            if word in title_text:
                subtype = word.capitalize()
                break

        # Rooms & area
        try:
            rooms = driver.find_element(By.XPATH, '//*[@id="main_content"]/div[8]/div/div[2]/div/div[1]/p').text.strip()
        except:
            rooms = "N/A"
        try:
            area = driver.find_element(By.XPATH, '//*[@id="main_content"]/div[8]/div/div[2]/div/div[2]/p').text.strip()
        except:
            area = "N/A"

        # Additional info (Kitchen, Terrace, Garden, etc.)
        main_container = driver.find_element(By.ID, "main_content")
        h3_elements = main_container.find_elements(By.TAG_NAME, "h3")
        keywords = ["Kitchen equipment", "Furnished", "Terrace", "Garden", "Number of facades", "Swimming pool", "State of the property"]
        info_dict = {k: "N/A" for k in keywords}

        for h3 in h3_elements:
            h3_text = h3.text.strip()
            if h3_text in ["Kitchen and bathrooms", "Outdoor description", "Interior description"]:
                try:
                    section_div = h3.find_element(By.XPATH, "following-sibling::div")
                    h4_elements = section_div.find_elements(By.TAG_NAME, "h4")
                    for h4 in h4_elements:
                        h4_text = h4.text.strip()
                        if h4_text in keywords:
                            try:
                                p_element = h4.find_element(By.XPATH, "following-sibling::p")
                                info_dict[h4_text] = p_element.text.strip()
                            except:
                                info_dict[h4_text] = "N/A"
                except:
                    pass

        # CSV'ye yaz
        writer.writerow([
            Property_ID, Locality_name, Postal_code, price,
            property_type, subtype,
            rooms, area,
            info_dict["Kitchen equipment"],
            info_dict["Furnished"],
            info_dict["Terrace"],
            info_dict["Garden"],
            info_dict["Number of facades"],
            info_dict["Swimming pool"],
            info_dict["State of the property"]
        ])
        print(f"Recorded: {Property_ID}")

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

def get_house_urls():
    driver = webdriver.Chrome()
    driver.get("https://immovlan.be/en/real-estate/house/for-sale")
    time.sleep(3)

    try:
        driver.find_element(By.ID, "didomi-notice-agree-button").click()
        time.sleep(2)
    except:
        pass

    ilanlar = driver.find_elements(By.XPATH, '//a[contains(@href, "/en/detail/")]')
    urls = [i.get_attribute("href") for i in ilanlar]
    driver.quit()
    return urls


def get_apartment_urls():
    driver = webdriver.Chrome()
    driver.get("https://immovlan.be/en/real-estate/apartment/for-sale")
    time.sleep(3)

    try:
        driver.find_element(By.ID, "didomi-notice-agree-button").click()
        time.sleep(2)
    except:
        pass

    ilanlar = driver.find_elements(By.XPATH, '//a[contains(@href, "/en/detail/")]')
    urls = [i.get_attribute("href") for i in ilanlar]
    driver.quit()
    return urls
```
